```python
def generate_category_frequency_table(df):
    freq_table = df['category'].value_counts().reset_index()
    freq_table.columns = ['Category', 'Count']
    return freq_table
def get_top_5_categories(df):
    freq_table = generate_category_frequency_table(df)
    return freq_table.head(5)
```

Measures of Centrality:

Calculate the mean, median, and mode for the price of products.
What's the average price point of products listed? How does this compare with the most common price point (mode)?
Measures of Dispersion:

Determine the variance, standard deviation, range, and interquartile range for product price.
How varied are the product prices? Are there any indicators of a significant spread in prices?
Visualizations:

Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. If its hard to read these diagrams, think why this is, and explain how it could be solved..
Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.


```python
from statistics import mean, median, mode
import pandas as pd

def calculate_central_tendency(df):
    prices = df['price'].dropna()
    return {
        'mean': prices.mean(),
        'median': prices.median(),
        'mode': prices.mode().tolist()  # mode can return multiple values
    }

def calculate_dispersion(df):
    prices = df['price'].dropna()
    return {
        'variance': prices.var(),
        'standard_deviation': prices.std(),
        'range': prices.max() - prices.min(),
        'interquartile_range': prices.quantile(0.75) - prices.quantile(0.25)
    }

import matplotlib.pyplot as plt  #Box Plot (for outliers)

def plot_price_histogram(df, bins=50):
    prices = df['price'].dropna()
    
    plt.figure(figsize=(10, 6))
    plt.hist(prices, bins=bins, edgecolor='black')
    plt.title("Histogram of Product Prices")
    plt.xlabel("Price ($)")
    plt.ylabel("Number of Products")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
central_tendency = calculate_central_tendency(df)
dispersion = calculate_dispersion(df)

print("Central Tendency:", central_tendency)
print("Dispersion Measures:", dispersion)

plot_price_histogram(df)
plot_price_boxplot(df)
```

Part 3: Unpacking Product Ratings
Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

Measures of Centrality:

Calculate the mean, median, and mode for the rating of products.
How do customers generally rate products? Is there a common trend?
Measures of Dispersion:

Determine the variance, standard deviation, and interquartile range for product rating.
Are the ratings consistent, or is there a wide variation in customer feedback?
Shape of the Distribution:

Calculate the skewness and kurtosis for the rating column.
Are the ratings normally distributed, or do they lean towards higher or lower values?
Visualizations:

Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python
def calculate_rating_central_tendency(df):
    ratings = df['stars'].dropna()
    return {
        'mean': ratings.mean(),
        'median': ratings.median(),
        'mode': ratings.mode().tolist()
    }
```

```python
def calculate_rating_dispersion(df):
    ratings = df['stars'].dropna()
    return {
        'variance': ratings.var(),
        'standard_deviation': ratings.std(),
        'interquartile_range': ratings.quantile(0.75) - ratings.quantile(0.25)
    }
```

```python
from scipy.stats import skew, kurtosis

def calculate_rating_distribution_shape(df):
    ratings = df['stars'].dropna()
    return {
        'skewness': skew(ratings),
        'kurtosis': kurtosis(ratings)
    }
    
```

```python
import matplotlib.pyplot as plt

def plot_rating_histogram(df, bins=10):
    ratings = df['stars'].dropna()

    plt.figure(figsize=(10, 6))
    plt.hist(ratings, bins=bins, edgecolor='black')
    plt.title("Distribution of Product Ratings")
    plt.xlabel("Rating (Stars)")
    plt.ylabel("Number of Products")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
rating_central = calculate_rating_central_tendency(df)
rating_dispersion = calculate_rating_dispersion(df)
rating_distribution_shape = calculate_rating_distribution_shape(df)

print("Central Tendency:", rating_central)
print("Dispersion:", rating_dispersion)
print("Distribution Shape:", rating_distribution_shape)
```
